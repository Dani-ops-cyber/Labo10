```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40,-10, 0, 8, 15, 22, 38], dtype=float)
farenheit =np.array([-40, 14, 32, 46, 59, 72, 100],dtype=float)
```

```python
capa=tf.keras.layers.Dense(units=1, input_shape=[1]) #le damos el input el numero de ntradas, units es la cantidad de neuronas
modelo=tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
historial=modelo.fit(celsius,farenheit, epochs=500, verbose=False)
```

```python
resultado=modelo.predict(np.array([100.0]))
print("Resultado: " + str(resultado[0][0]) + " Fahrenheit")
```
