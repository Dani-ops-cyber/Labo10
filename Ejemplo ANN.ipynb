**Artificial Neural Network o Red Neuronal Artificial**

un modelo de procesamiento inspirado en el funcionamiento del cerebro humano, utilizado para identificar patrones y realizar tareas de clasificación en imágenes y videos. Las redes neuronales artificiales consisten en capas de neuronas interconectadas que procesan datos de manera jerárquica, donde cada capa extrae características de nivel superior de los datos anteriores.

**APLICACION DE ANN**


Clasificación de Imágenes: Identificar y clasificar objetos en imágenes, como determinar si una imagen contiene un perro, un gato, o un coche.
Detección de Objetos: Localizar y etiquetar objetos específicos dentro de una imagen o video.
Segmentación de Imágenes: Dividir una imagen en diferentes partes o regiones para identificar objetos y sus límites.
Reconocimiento Facial: Identificar rostros específicos en imágenes o secuencias de video.
Reconocimiento de Patrones y Texturas: Identificar patrones o texturas específicas en una imagen, útil en industrias como la de textiles o inspección de calidad.
ANNs también pueden ser parte de arquitecturas más complejas como las redes convolucionales (CNN), que son muy eficaces para analizar datos visuales, ya que están diseñadas para procesar y extraer información espacial de las imágenes.

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
```

Realizaremos un modelo sencillo para determinar el correcto funcionamiento de la ecuación y=3x + 2, para ello generaremos patrones numéricos según la ecuación

```python
xs = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
ys = np.array([-1, 2, 5, 8, 11, 14], dtype=float)
```

generaremos un modelo secuencial porque vamos a especificar cada capa del modelo de forma secuencial. Secuencialmente significa que el modelo tendrá una entrada y una salida con las que trabajar.

```python
model = keras.Sequential()
```

```python
model.add(layers.Dense(3, input_dim=1)) #First 1 layer, then 2 (5-3) and finally 2 (3-3)
model.add(layers.Dense(3))
model.add(layers.Dense(1))
```

```python
model.compile(keras.optimizers.Adam(0.02), loss="mean_squared_error", metrics=['accuracy'])
```

```python
model.fit(xs, ys, epochs=100)
```

```python
predictions = model.predict(xs)
rounded_predictions = np.round(predictions)
_, accuracy = model.evaluate(xs, ys)
print("The accuracy is {}".format(accuracy))
```
