Una tienda veterinaria quiere determinar si deberían comenzar a vender “tarjetas de tienda veterinaria” para sus
compras de los clientes. Con esta finalidad dispone del siguiente conjunto de datos:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
clients = pd.DataFrame(
    {
        "costo_pro": [
            750,
            1245,
            230,
            533,
            490,
            1000,
            190,
            900,
            600,
            50,
            1100,
            930,
            450,
            330,
            750

        ],
        "num_compra_cre": [
            3,
            1,
            4,
            3,
            2,
            1,
            0,
            3,
            2,
            1,
            0,
            4,
            3,
            2,
            0
        ],
    }
)
```

```python
scaler = MinMaxScaler().fit(clients.values)
clients = pd.DataFrame(
    scaler.transform(clients.values), columns=["costo_pro", "num_compra_cre"]
)
print(clients)
```

```python
#Entrenamiento
kmeans = KMeans(n_clusters=5).fit(clients.values)
print(kmeans.labels_)
```

```python
#Encontramos que el entramiento agrupo en dos grupos los valores, 0 y 1 clasificandolos asi
```

```python
#Segun el nro de clusters se divide en esa cantidad de grupos, segun el conjunto de datos
```

```python
clients["cluster"] = kmeans.labels_
print(kmeans.cluster_centers_, kmeans.inertia_)
```

```python
plt.figure(figsize=(6, 5), dpi=100)
colors = ["red", "orange", "blue", "black", "purple", "pink", "brown"]
print(clients)
for cluster in range(kmeans.n_clusters):
    plt.scatter(
        clients[clients["cluster"] == cluster]["costo_pro"],  # Pandas dataframe filter
        clients[clients["cluster"] == cluster][
            "num_compra_cre"
        ],  # Pandas dataframe filter
        marker="x",
        s=180,
        color=colors[cluster],
        alpha=0.5,
    )
    plt.scatter(
        kmeans.cluster_centers_[cluster][0],
        kmeans.cluster_centers_[cluster][1],
        marker="o",
        s=280,
        color=colors[cluster],
    )
plt.title("Clientes")
plt.xlabel("Costo promedio de compra (bs)")
plt.ylabel("promedio de compras con crédito tarjeta")
plt.text(1.1, 0.5, "K={}".format(kmeans.n_clusters))
plt.text(1.1, 0.3, "Inertia={}".format(round(kmeans.inertia_,2)))
plt.show()
```
