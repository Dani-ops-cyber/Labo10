¿Qué es la función de pérdida? Describa algunas de las funciones de pérdida existentes.

La función de pérdida mide la diferencia entre las predicciones del modelo y los valores reales. Es la métrica que el optimizador intenta minimizar. Una función de pérdida alta significa que las predicciones del modelo son incorrectas en mayor medida, mientras que una pérdida baja indica que las predicciones se acercan a los valores reales.

MSE (Mean Squared Error): Calcula el promedio de los cuadrados de los errores (diferencias) entre las predicciones y los valores reales. Es común en problemas de regresión.

Cross-Entropy (Entropía cruzada): Mide la distancia entre dos distribuciones de probabilidad y se usa frecuentemente en problemas de clasificación, especialmente en redes neuronales.

MAE (Mean Absolute Error): Calcula el promedio de los errores absolutos entre las predicciones y los valores reales, útil en problemas donde queremos reducir el impacto de errores grandes.

Hinge Loss: Usada en máquinas de soporte vectorial (SVM), maximiza el margen entre las clases en un problema de clasificación.


# ** Mean Squared Error (MSE)**

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

xs = np.array([3.3, 5], dtype=float)
ys = np.array([11.9, 17], dtype=float)


model1 = keras.Sequential()
model1.add(layers.Dense(4, input_dim=1))  # Primera capa una neurona
model1.add(layers.Dense(4))               # Segunda capa
model1.add(layers.Dense(4))               # Tercera capa
model1.add(layers.Dense(1))               # Capa de salida
```

```python
model1.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model1.fit(xs, ys, epochs=300)
```

```python
predictions = model1.predict(xs)
rounded_predictions = np.round(predictions)
_, accuracy = model1.evaluate(xs, ys)
print("The accuracy is {}".format(accuracy))
```

```python
resultado=model1.predict(np.array([10.0]))
print("Resultado Y: " + str(resultado[0][0]))
```

# **Mean Absolute Error (MAE)*

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

xs = np.array([3.3, 5], dtype=float)
ys = np.array([11.9, 17], dtype=float)


model2 = keras.Sequential()
model2.add(layers.Dense(4, input_dim=1))  # Primera capa una neurona
model2.add(layers.Dense(4))               # Segunda capa
model2.add(layers.Dense(4))               # Tercera capa
model2.add(layers.Dense(1))               # Capa de salida
```

```python
# Compilar el modelo usando el optimizador Adam
model2.compile(optimizer=keras.optimizers.Adam(learning_rate=0.1), loss="mean_squared_error", metrics=['accuracy'])
model2.fit(xs, ys, epochs=300)
```

```python
predictions = model2.predict(xs)
rounded_predictions = np.round(predictions)
_, accuracy = model2.evaluate(xs, ys)
print("The accuracy is {}".format(accuracy))
```

```python
resultado=model2.predict(np.array([10.0]))
print("Resultado Y: " + str(resultado[0][0]))
```

# ** Huber Loss**

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

xs = np.array([3.3, 5], dtype=float)
ys = np.array([11.9, 17], dtype=float)


model3 = keras.Sequential()
model3.add(layers.Dense(4, input_dim=1))  # Primera capa una neurona
model3.add(layers.Dense(4))               # Segunda capa
model3.add(layers.Dense(4))               # Tercera capa
model3.add(layers.Dense(1))               # Capa de salida
```

```python
# Compilar el modelo usando Huber Loss como función de pérdida
model3.compile(optimizer='adam', loss=keras.losses.Huber(), metrics=['accuracy'])
model3.fit(xs, ys, epochs=30)

```

```python
predictions = model3.predict(xs)
rounded_predictions = np.round(predictions)
_, accuracy = model3.evaluate(xs, ys)
print("The accuracy is {}".format(accuracy))
```

```python
resultado=model3.predict(np.array([10.0]))
print("Resultado Y: " + str(resultado[0][0]))
```
