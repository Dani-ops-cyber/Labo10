```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import pandas as pd

# Cargar datos desde un archivo CSV
df = pd.read_csv('/content/drive/MyDrive/embebidos_II_Apuntes/Vision artificial con EMbebidos/spam text data.csv')  # Cambia 'data.csv' por la ruta de tu archivo

# Vectorización del texto
vectorizer = CountVectorizer(stop_words='english')

# Convertimos los mensajes en una matriz de características
X = vectorizer.fit_transform(df['Message'])

# Convertimos las etiquetas en valores numéricos: "ham" = 0, "spam" = 1
y = df['Category'].map({'ham': 0, 'spam': 1})

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=50)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Crear el modelo
model = MultinomialNB()

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Nuevos mensajes
nuevos_mensajes = ["Congratulations! You have won a prize!", "Can we schedule a meeting for tomorrow?"]

# Vectorizar los nuevos mensajes
X_nuevos = vectorizer.transform(nuevos_mensajes)

# Predecir si son spam o no
predicciones = model.predict(X_nuevos)

# Mostrar los resultados
for mensaje, prediccion in zip(nuevos_mensajes, predicciones):
    print(f'Mensaje: {mensaje} - Predicción: {"spam" if prediccion == 1 else "ham"}')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Usar TF-IDF para la vectorización
vectorizer_tfidf = TfidfVectorizer(stop_words='english')

X_tfidf = vectorizer_tfidf.fit_transform(df['Message'])

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.5, random_state=42)

# Entrenar el modelo
model.fit(X_train, y_train)

# Evaluar
y_pred = model.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
# Nuevos mensajes
nuevos = ["¡Felicidades! Has ganado un premio!", "¿Podemos programar una reunión para mañana?"]

# Vectorizar los nuevos mensajes
X_nuevosi = vectorizer.transform(nuevos)

# Predecir si son spam o no
prediccionesi = model.predict(X_nuevosi)

# Mostrar los resultados
for mensaje, prediccion in zip(nuevos, prediccionesi):
    print(f'Mensaje: {mensaje} - Predicción: {"spam" if prediccion == 1 else "ham"}')
```
